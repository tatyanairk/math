#### Напишите код, который будет рисовать график окружности в полярных координатах.

```python
def circle_(R):
    x = np.linspace(0, 5, 100)
    y = np.sqrt(R**2 - x**2)
    plt.polar(x, y)
```

#### y = x2 – 1
#### exp(x) + x∙(1 – y) > 1

```python
import numpy as np
from scipy.optimize import fsolve
def inequat(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) - (1 - y)*x - 1)
x1, y1 = fsolve(inequat, (1, 1))
if np.exp(x1) - (1 - y1)*x1 - 1 > 0:
    print (x1, y1)
else:
    print("Нет решения")
```
