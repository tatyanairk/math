#### task 5.1

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = 2*X + 5*Y
ax.plot_wireframe(X, Y, Z)
Z1 = 2*X + 5*Y + 5
ax.plot_wireframe(X, Y, Z1)
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2
ax.plot_wireframe(X, Y, Z)
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = X**2 - Y**2
ax.plot_wireframe(X, Y, Z)
```
